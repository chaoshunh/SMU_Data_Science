# 1970 Draft Distribution

```python
import pandas as pd
import numpy as np

df = pd.read_excel('1970_draft_numbers.xlsx')
df.head()
```

```python
df.ix[:,1:].T.values.reshape(-1)
```

```python
df.values # gives arrays in order of days per month, eg first array is first day of each month's valu
# transpose to move current values in columns to rows and create a singular array
# this inverts the days to columns, 31 columns in this case

df.values.T.shape
```

```python
df.ix[:,1:].values.T.reshape(-1) # reshape strips dimensionality back 1 if -1 is used
```

```python
# create a single array in order of days
# transpose days which go down columns to go across rows in order to get values for entire month
draft1970 = [i for i in df.ix[:,1:].values.T.reshape(-1) if ~np.isnan(i)]
np.corrcoef(range(len(draft1970)), draft1970)
```

```python
draft1970
```

```python
import seaborn as sns

from matplotlib import pyplot as plt

%matplotlib inline

plt.plot(list(range(len(draft1970))), draft1970)
plt.xlim(0, 366)
```

## Means and medians for each month

```python
df
```

```python
means1970 = df.apply(lambda x: x.mean(), axis=0)
print(means1970)
plt.plot(means1970)
plt.ylim(0, 250)
```

```python
medians1970 = df.apply(lambda x: x.median(), axis=0)
print(medians1970)
plt.plot(medians1970)
plt.ylim(0, 275)
```

# 1971 Draft Distribution

```python
df1971 = pd.read_excel('/users/bmwes/onedrive/documents/school/statistical sampling/lab2/1971_draft_numbers2.xlsx')
df1971.head()
```

```python
draft1971 = [i for i in df1971.values.T.reshape(-1) if not np.isnan(i)]
np.corrcoef(list(range(len(draft1971))), list(draft1971))
```

```python
import seaborn as sns

from matplotlib import pyplot as plt

%matplotlib inline

plt.plot(list(range(len(draft1971))), draft1971)
plt.xlim(0, 366)
```

## Means and medians for 1971

```python
means1971 = df1971.apply(lambda x: x.mean(), axis=0)
print(means1971)
plt.plot(list(range(df1971.shape[1])), means1971)
plt.ylim(0, 300)
```

```python
medians1971 = df1971.apply(lambda x: x.mean(), axis=0)
print(medians1971)
plt.plot(list(range(df1971.shape[1])), means1971)
plt.ylim(0, 300)
```
