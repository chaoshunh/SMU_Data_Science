## Exploratory Data Analysis

A chi-squared test of independence on our categorical features shows that all are significantly related to our dependent variable Default. The most dependent variable is sub-grade which supports that ultimate disposition of the loan is highly correlated with the proprietary Lending Club grade and makes intuitive sense. Effect size, or strength of association, is also strongest (but considered moderately strong) between sub-grade and default at 0.12, which is only a "small" to "medium" effect. Income verification_status has almost no association with the default disposition. Of the categorical features, sub-grade would be our best predictor. 

```python
import scipy

cols = ['sub_grade','home_ownership','verification_status','purpose']

def chisqr(df, features):
    vals = []
    for i in features:
        cross = pd.crosstab(df[i], df['default'])
        chisq = scipy.stats.chi2_contingency(cross)
        cramer = np.sqrt((chisq[0]/len(df)) / np.minimum(len(df)-1,len(cols)-1))
        chisq = list(chisq[0:2])
        chisq.append(cramer)
        vals.append(chisq)
    return DataFrame(vals)

chi2 = u'\N{GREEK SMALL LETTER CHI}\N{SUPERSCRIPT TWO}'
plt.figure(figsize=(10,10))
hm = sns.heatmap(chisqr(df,cols),cbar=True,annot=True,square=True, fmt='.2f',annot_kws={'size':15},
                 yticklabels=cols,xticklabels=[chi2 + ' Score','p-value','Cramer\'s V'])
plt.show()
```

The pairplot below shows loan amount, total accounts and open accounts are all skewed to the right. Annual income has several outliers, which we mitigated via the log transformation in our "Outliers" section.

## Logistic Regression

For training and testing purposes, we separated 80% of the instances for training and the remaining 20% for testing. We repeated this process of separating the testing and training data ten times. We will use the hold out cross validation method built into scikit-learn.

```python
if 'default' in df:
    y = df['default'].values 
    del df['default']
    X = df.values 
num_cv_iterations = 10
num_instances = len(y)
cv_object = ShuffleSplit(n=num_instances, n_iter=num_cv_iterations, test_size  = 0.2)
print (cv_object)
```

To thoroughly assess model performance, the model accuracy measurement and confusion matrix are used. Below, we compared L1 and L2 regularization. 

```python
from sklearn.linear_model import LogisticRegression
penalty = ['l1', 'l2']

for p in penalty:
    print ('\nRegularization:', p)
    print('Class Weight: None')
    lr_clf = LogisticRegression(penalty=p, C=1, class_weight= None) 
    iter_num = 0        
    for train_indices, test_indices in cv_object:
        X_train = X[train_indices]
        y_train = y[train_indices]
        X_test = X[test_indices]
        y_test = y[test_indices]
        lr_clf.fit(X_train,y_train)  
        y_hat = lr_clf.predict(X_test) 
        
        accuracy = mt.accuracy_score(y_test,y_hat)
        confusion_matrix = mt.confusion_matrix(y_test,y_hat)
        sensitivity = mt.recall_score(y_test, y_hat)
        specificity = confusion_matrix[0,0]/float(confusion_matrix[0,1]+confusion_matrix[0,0])
        precision = mt.precision_score(y_test, y_hat) 
        print ("====Iteration",iter_num," ====")
        print ('Accuracy: %.2f' %accuracy)
        print ('confusion matrix',confusion_matrix)
       
        iter_num+=1
    
    print ('Average Accuracy Score: %.2f' %np.mean(accuracy)) 
    print ('Average Sensitivity Score: %.2f' %np.mean(sensitivity)) 
    print ('Average Specificity Score: %.2f' %np.mean(specificity)) 
    print ('Average Precision Score: %.2f' %np.mean(precision))     
```

Assigning class weight equal to None, the average accuracy of the ten iterations using both L1 and L2 regularizations is almost equal at 85%. Even though both have specificity of 100%, their sensitivity score is negiligible (0.01% for L1 and 0% for L2). The average precision scores of both L1 and L2 are also far below optimum. We measure our model performance based on how well it predicts the 'default' class. The above model is usless in this regard. This calls for attempting a better model. 

The binary output variable is highly unbalanced (15% one and 85% zero); therefore, we should increase the weight of class 1 relative to class 0 to mitigate the impact of this potential issue.

```python
penalty = ['l1', 'l2']

for p in penalty:
    print ('Regularization:', p)
    print('Class Weight: Balanced')
    lr_clf = LogisticRegression(penalty=p, C=1, class_weight= 'balanced') 
    iter_num = 0        
    for train_indices, test_indices in cv_object:
        X_train = X[train_indices]
        y_train = y[train_indices]
        X_test = X[test_indices]
        y_test = y[test_indices]
        lr_clf.fit(X_train,y_train)  
        y_hat = lr_clf.predict(X_test) 
        
        accuracy = mt.accuracy_score(y_test,y_hat)
        confusion_matrix = mt.confusion_matrix(y_test,y_hat)
        sensitivity = mt.recall_score(y_test, y_hat)
        specificity = confusion_matrix[0,0]/float(confusion_matrix[0,1]+confusion_matrix[0,0])
        precision = mt.precision_score(y_test, y_hat) 
        print ("====Iteration",iter_num," ====")
        print ('Accuracy: %.2f' %accuracy)
        print ('confusion matrix',confusion_matrix)
       
        iter_num+=1
    
    print ('Average Accuracy Score: %.2f' %np.mean(accuracy)) #Average accuracy of the three iteration
    print ('Average Sensitivity Score: %.2f' %np.mean(sensitivity)) #Average sensitivity of the three iteration
    print ('Average Specificity Score: %.2f' %np.mean(specificity)) #Average specificity of the three iteration
    print ('Average Precision Score: %.2f' %np.mean(precision)) #Average precision of the three iteration 
```

Considering the class imbalance in the output variable produces signficant differences in our model predictions. The average accuracy and specificity scores reduced to 65% for both L1 and L2 regularization. On the contrary, the average sensitivity score increased to 65% for both regularization techniques. The precision score has not changed. We did not observe a significant difference in these scores between L1 and L2 forms. Moreover, there is no significant difference in the scores among the various iterations. Even though the sensitivity score has increased, the models still have very low precision. Therefore, these models are far from optimum. We will attempt to change the cost variable and see if the model performance improves.

<a id='pipeline'></a>
### Pipeline

```python
from sklearn.pipeline import Pipeline

std_scl = StandardScaler()
lr_clf = LogisticRegression(penalty='l2', C=1, class_weight='balanced') 

# create the pipline
piped_object = Pipeline([('scale', std_scl), ('logit_model', lr_clf)])

# run the pipline crossvalidated
for iter_num, (train_indices, test_indices) in enumerate(cv_object):
    piped_object.fit(X[train_indices],y[train_indices])  # train object
    

trained_model_from_pipeline = piped_object.named_steps['logit_model']

# now look at the weights
weights = pd.Series(trained_model_from_pipeline.coef_[0],index=df.columns)
weights.plot(kind='bar')
plt.show()
```

We can again interpret the output from our pipeline and our scaled variables using the coefficient of annual income as an example. Since the coefficient of annual income is negative (and once again the most impactful of all of our weights), we can say that an increase in annual income is associated with a decrease in the odds of loan default.  Aside from interest rate (which we feel is an effect and not a cause), the coefficient of inquiries last 6 months is positive and the most significant of our continuous variables.  This means that an increase in the number of inquiries over the most recent six-month period is associated with an increase in the odds of loan default.  Both of these observations are also intuitive.

### Third Degree Polynomial Kernel

<a id='sigmoid_kernel'></a>
### Sigmoid kernel

```python
svm_clf_sig = SVC(kernel='sigmoid')
svm_clf_sig.fit(X_train_scaled, y_train)

y_hat_sig = svm_clf_sig.predict(X_test_scaled)

acc_sig = mt.accuracy_score(y_test, y_hat_sig)
conf_sig = mt.confusion_matrix(y_test, y_hat_sig)
print('Accuracy:', acc_sig)
print(conf_sig)
```

When training SVM models using four different kernels (linear, third degree polynomial, rbf, and sigmoid), we get a relatively high accuracy.  However, when taking a closer look at the confusion matrices for each of these models it is apparent that none of them are actually performing well.  Each model has lots of false negatives.  In fact, the third degree polynomial model is the only one that didn't just predict 0 for every single sample.  Since this kernel seemed to perform slightly better than the others (though it admittedly still performed incredibly poorly), we will continue to tweak this model to see if it's possible to get more useful results.

Given that our response value is highly skewed, the first thing we will do is change the weights.  We initially used a default weighting of 1 for each value.  We will try using the 'balanced' weighting mode instead, which uses the values of the response variable to automatically adjust the weights.  The adjusted weights are inversely proportional to the class frequencies in the input data.

```python
svm_clf_poly = SVC(kernel='poly', degree=3, class_weight='balanced')
svm_clf_poly.fit(X_train_scaled, y_train)

y_hat_poly = svm_clf_poly.predict(X_test_scaled)

acc_poly = mt.accuracy_score(y_test, y_hat_poly)
conf_poly = mt.confusion_matrix(y_test, y_hat_poly)
print('Accuracy:', acc_poly)
print(conf_poly)
```

When using balanced class weights, we get a lower accuracy, but a much more useful model.  We now have fewer false negatives and also more true positives.  So we are now predicting some defaulted loans, which is the whole point of the model that we are building.  

However, we also have a lot of false positives now.  These are not as concerning as false negatives in this case, because we would rather incorrectly predict that a loan will default than incorrectly predict that it won't default.  We will do some more tweaking to see if this model can be improved some more.  The next thing we will do is try multiple different values for the penalty parameter C.

```python
for c in [0.1, 0.3, 1, 3, 10]:
    svm_clf_poly = SVC(C=c, kernel='poly', degree=3, class_weight='balanced')
    svm_clf_poly.fit(X_train_scaled, y_train)

    y_hat_poly = svm_clf_poly.predict(X_test_scaled)

    acc_poly = mt.accuracy_score(y_test, y_hat_poly)
    conf_poly = mt.confusion_matrix(y_test, y_hat_poly)
    print('--- Cost = ' + str(c) + ' ---')
    print('Accuracy:', acc_poly)
    print(conf_poly)
```

After training the model with costs of 0.1, 0.3, 1, 3, and 10 it appears that 0.3 is the optimal cost out of these options.  The accuracy dips slightly to 68% with a 0.3 cost as opposed to 70% with a 0.1 cost, but we also see a significant bump in true positive classifications.  With costs of 1, 3, and 10 we see about the same level of accuracy but we see diminishing returns in terms of true positive classifications.  We did attempt to evaluate costs of 30 and 100, but observed little benefit for the computational investment.
